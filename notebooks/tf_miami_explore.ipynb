```python
main_df = pd.read_csv('miami-housing.csv')
coor_x = "LATITUDE"
coor_y = "LONGITUDE"
label_altitudes = [
    "SALE_PRC",
    "LND_SQFOOT",
    "TOT_LVG_AREA",
    "SPEC_FEAT_VAL",
]
for label_altitude in label_altitudes:
    plotted_df = main_df[[coor_x,coor_y,label_altitude]].copy()
    plotted_df = plotted_df.dropna()
    draw_map(
        input_df=plotted_df,
        coor_x=coor_x,
        coor_y=coor_y,
        label=label_altitude,
        group_by="mean",
        scale_factor=500,
        # log=True,
    )
```

```python
main_df = pd.read_csv('beijing-30k.csv')
coor_x = "Lat"
coor_y = "Lng"
label_altitudes = [
    "price",
    "totalPrice",
    "communityAverage",
]
for label_altitude in label_altitudes:
    plotted_df = main_df[[coor_x,coor_y,label_altitude]].copy()
    plotted_df = plotted_df.dropna()
    draw_map(
        input_df=plotted_df,
        coor_x=coor_x,
        coor_y=coor_y,
        label=label_altitude,
        group_by="mean",
        scale_factor=500,
        # log=True,
    )
```

```python
main_df = pd.read_csv('beijing-30k.csv')
coor_x = "Lat"
coor_y = "Lng"
label_altitude = "constructionTime"
plotted_df = main_df[[coor_x,coor_y,label_altitude]].copy()
plotted_df = plotted_df.dropna()
plotted_df[label_altitude] = plotted_df[label_altitude] - 1940
draw_map(
    input_df=plotted_df,
    coor_x=coor_x,
    coor_y=coor_y,
    label=label_altitude,
    group_by="mean",
    scale_factor=500,
    # log=True,
)
```
